```python
import pandas as pd

# Membaca data dari file Excel
df = pd.read_excel("data_encode.xlsx")

# Mengubah data menjadi list kategori buku berdasarkan Tgl pinjam
kategori_buku = []
for _, row in df.iterrows():
    kategori = []
    if row["Agama"] == 1:
        kategori.append("Agama")
    if row["Anak"] == 1:
        kategori.append("Anak")
    if row["Biografi"] == 1:
        kategori.append("Biografi")
    if row["Dongeng"] == 1:
        kategori.append("Dongeng")
    if row["Komik"] == 1:
        kategori.append("Komik")
    if row["Legenda"] == 1:
        kategori.append("Legenda")
    if row["Novel"] == 1:
        kategori.append("Novel")
    if row["Pendidikan"] == 1:
        kategori.append("Pendidikan")
    if row["Pengetahuan"] == 1:
        kategori.append("Pengetahuan")
    
    kategori_buku.append(kategori)

# Menampilkan hasil dengan format yang diberikan
for kategori in kategori_buku:
    print(kategori)
```

```python
from itertools import chain
from collections import Counter

# Menggabungkan semua kategori buku menjadi satu list
all_categories = list(chain(*kategori_buku))

# Menghitung kemunculan setiap kategori
kategori_counts = Counter(all_categories)

# Menampilkan kandidat 1 item (C1)
c1 = list(kategori_counts.keys())
print(c1)
```

```python
from collections import Counter

# Menghitung frekuensi masing-masing item dari C1
frekuensi_item = Counter(all_categories)

# Menampilkan jumlah frekuensi masing-masing item
for item, frekuensi in frekuensi_item.items():
    print(f"Kategori: {item}, Frekuensi: {frekuensi}")
```

```python
from collections import Counter

# Menghitung frekuensi masing-masing item dari C1
frekuensi_item = Counter(all_categories)

# Menghitung jumlah total transaksi
total_transaksi = len(kategori_buku)

# Menghitung nilai support item
support_item = {}
for item, frekuensi in frekuensi_item.items():
    support = frekuensi / total_transaksi
    support_item[item] = support

# Menampilkan nilai support item
for item, support in support_item.items():
    print(f"Item: {item}, Support: {support}")
```

```python
# Nilai minimum support
minimum_support = 0.02

# Menentukan large 1-itemset
large_1_itemset = []
for item, support in support_item.items():
    if support >= minimum_support:
        large_1_itemset.append((item, support))

# Menampilkan large 1-itemset
for item, support in large_1_itemset:
    print(f"Item: {item}, Support: {support}")
```

```python
from itertools import combinations

# Membuat kandidat 2-itemset dari kombinasi large 1-itemset
kandidat_2_itemset = list(combinations([item for item, _ in large_1_itemset], 2))

# Menampilkan kandidat 2-itemset
for itemset in kandidat_2_itemset:
    print(itemset)
```

```python
from collections import defaultdict

# Menginisialisasi dictionary untuk menghitung frekuensi kombinasi 2-itemset
frekuensi_kombinasi = defaultdict(int)

# Menghitung frekuensi kombinasi 2-itemset dalam setiap transaksi
for transaksi in kategori_buku:
    for kombinasi in kandidat_2_itemset:
        if all(item in transaksi for item in kombinasi):
            frekuensi_kombinasi[kombinasi] += 1

# Menampilkan jumlah frekuensi kombinasi 2-itemset
for kombinasi, frekuensi in frekuensi_kombinasi.items():
    print(f"Kombinasi: {kombinasi}, Frekuensi: {frekuensi}")
```
